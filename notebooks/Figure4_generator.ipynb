```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes
```

```python
data = pd.read_csv("figure_3.csv")
data
```

```python
dataC = data.copy()
residue_cutoff = 699
residue_cutoff_top = 906

filtered_dataC = dataC[(dataC['residueC'] >= residue_cutoff) & (dataC['residueC'] < residue_cutoff_top)]
filtered_dataC.reset_index(drop=True, inplace=True)

plt.rcParams['font.size'] = 10
fig, (ax2, ax1) = plt.subplots(2, 1, sharex=True, figsize=(2, 2))
fig.subplots_adjust(hspace=0.3)

sns.barplot(data=filtered_dataC, x='residueC', y='energyC', color='k', ax=ax2, linewidth=5)
sns.barplot(data=filtered_dataC, x='residueC', y='energyC', color='k', ax=ax1, linewidth=5)

widths = 5

for patch in ax1.patches:
    patch.set_width(widths)  
for patch in ax2.patches:
    patch.set_width(widths) 

ax1.set_ylim(-9, -7)
ax2.set_ylim(-1, filtered_dataC['energyC'].max() + 1)
ax1.set_xlim(filtered_dataC.index.min(), filtered_dataC.index.max())
ax2.set_xlim(filtered_dataC.index.min(), filtered_dataC.index.max())
ax2.spines.bottom.set_visible(False)
ax1.spines.top.set_visible(False)

ax2.spines.right.set_visible(False)
ax1.spines.right.set_visible(False)
ax2.spines.top.set_visible(False)

ax2.tick_params(labelbottom=False, bottom=False)
ax1.set_xticks([1, 51, 101, 151, 201])
ax1.set_xticklabels(['700','', '800', '', '900'])
ax2.set_yticks([-1, 0, 2])
ax1.set_yticks([-7, -8, -9])


d = .5
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12, linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax2.plot([0, 0], [0, 0], transform=ax2.transAxes, **kwargs)
ax1.plot([0, 0], [1, 1], transform=ax1.transAxes, **kwargs)


ax1.set_xlabel('')
ax1.set_ylabel('')
ax2.set_xlabel('')
ax2.set_ylabel('')
fig1 = plt.gcf()
fig1.savefig('figure_3_c.png', bbox_inches='tight', pad_inches=0, transparent=True)
plt.show()
```

```python
dataD = data.copy()
residue_cutoff = 699
residue_cutoff_top = 906

filtered_dataD = dataD[(dataD['residueD'] >= residue_cutoff) & (dataD['residueD'] < residue_cutoff_top)]
filtered_dataD.reset_index(drop=True, inplace=True)
plt.rcParams['font.size'] = 10

fig, (ax2, ax1) = plt.subplots(2, 1, sharex=True, figsize=(2, 2))
fig.subplots_adjust(hspace=0.3)  
sns.barplot(data=filtered_dataD, x='residueD', y='energyD', color='k', ax=ax2)
sns.barplot(data=filtered_dataD, x='residueD', y='energyD', color='k', ax=ax1)

ax2.set_ylim(7, 9)
ax1.set_ylim(-2, 1)   

ax1.set_xlim(filtered_dataD.index.min(), filtered_dataD.index.max())  
ax2.set_xlim(filtered_dataD.index.min(), filtered_dataD.index.max())  

ax2.spines.bottom.set_visible(False)
ax1.spines.top.set_visible(False)


ax2.tick_params(labelbottom=False, bottom=False)
ax2.set_yticks([7, 8, 9])
ax1.set_xticks([1, 51, 101, 151, 201])
ax1.set_xticklabels(['700', '', '800', '', '900'])

ax2.spines.right.set_visible(False)
ax1.spines.right.set_visible(False)
ax2.spines.top.set_visible(False)

ax1.set_yticks([-2, -1, 0, 1])
widths = 5

for patch in ax1.patches:
    patch.set_width(widths) 

for patch in ax2.patches:
    patch.set_width(widths)  


d = .5  
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax2.plot([0, 0], [0, 0], transform=ax2.transAxes, **kwargs)
ax1.plot([0, 0], [1, 1], transform=ax1.transAxes, **kwargs)

ax1.set_xlabel('')
ax1.set_ylabel('')
ax2.set_xlabel('')
ax2.set_ylabel('')
fig1 = plt.gcf()
fig1.savefig('figure_3_d.png', bbox_inches='tight', pad_inches=0, transparent=True)
plt.show()
```
